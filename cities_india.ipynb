```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code != 200:
    print("Failed to retrieve the webpage.")
    exit()

# Use pandas to read all tables from the page
tables = pd.read_html(response.text)

# Check if any tables were found
if not tables:
    print("No tables found on the page.")
    exit()



```

```python
df = tables[0]
```

```python
df.columns
```

```python
df = df[['City','State or union territory', 'Population (2011)[3]']]
```

```python
df = df.rename(columns={'Population (2011)[3]': 'Population (2011)'})

df
```

```python
df1 = tables[1]
df1.columns
```

```python
df1 = df1[['City','State or union territory', 'Population (2011)[5]']]
df1 = df1.rename(columns={'Population (2011)[5]': 'Population (2011)'})
df1
```

```python
merged_df = pd.concat([df, df1], ignore_index=True)
merged_df
```
